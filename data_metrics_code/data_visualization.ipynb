```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../results/peter_pan_emotions_041021_232002.csv")
```

```python
df['emotions'] = df[['joy','sadness','anger','love','thankfulness','fear']].idxmax(axis=1)
```

```python
df
```

```python
df['count'] = df["emotions"].groupby(df["emotions"]).transform('count')
```

```python
df.head()
```

```python
emotion_name = df['emotions'].unique()
emotion_count = df['count'].unique()
```

```python
plt.figure(figsize=(10,5))
g = sns.barplot(x=emotion_name, y=emotion_count)
```
