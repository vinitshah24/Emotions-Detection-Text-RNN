```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("results/detected_emotions_output_031521_235302.csv")
```

```python
df
```

```python
df['emotion'] = df[['joy','sadness','anger','love','thankfulness','fear']].idxmax(axis=1)
```

```python
df
```

```python
df['count'] = df.emotion.groupby(df.emotion).transform('count')
```

```python
df.head()
```

```python
emotion_name = df['emotion'].unique()
emotion_count = df['count'].unique()
```

```python
plt.figure(figsize=(10,5))
g = sns.barplot(x=emotion_name, y=emotion_count)
```
